```python
# @title Task 10
"""
Create features list including variables from var_0 to var_9.
a. Find correlations in features.
b. Sort most correlated variables in features.
c. Show top 10 most correlated variables in features.
"""
features = []
for i in range(10):
  features.append('var_' + str(i))
correlations = df[features].corr()
correlations_absolute = correlations.abs()      # абсолютные значения корреляций
correlations_unstacked = correlations_absolute.unstack()      # таблица в плоский список (индекс становится колонкой)
correlations_sorted = correlations_unstacked.sort_values(ascending=False)     # сортировка значений по убыванию
correlations = correlations_sorted[correlations_sorted != 1]
print(correlations.head(10))
```

```python
# @title Task 11
"""
Remove outliers from dataset.
"""
def remove_outliers(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
  return df_filtered

df_no_outliers = remove_outliers(df, 'var_0')
df_no_outliers
```
