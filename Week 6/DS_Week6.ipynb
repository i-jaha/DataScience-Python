## Ridge Regression

```python
# Implement and see the effect of Ridge Regression
# using plots of different Ridge regression lines

# Import Pandas for data processing
import pandas as pd
```

```python
# Read the CSV file
dataset = pd.read_csv('ridge.csv')
df = dataset.copy()
```

```python
# Split into X (Independent) and Y (predicted)
X = df.iloc[:, :-1]
Y = df.iloc[:,  -1]
```

```python
# Import Ridge and matplotlib
from sklearn.linear_model import Ridge
import matplotlib.pyplot as plt
```

```python
# Create test data for plotting
X_plt = [0,1,2,3,4]

# Create a list of different alpha/penalty parameter values
ridge_l = [0,1,10,100]
```

```python
# Plot different Ridge Regression lines in one figure
for i, l in enumerate(ridge_l):
    ridge = Ridge(alpha=l)
    ridge.fit(X, Y)

    ridge_coeff = ridge.coef_
    ridge_intercept = ridge.intercept_
    Y_plt = ridge.predict(pd.DataFrame(X_plt))

    plt.figure(2)
    plt.plot(X_plt, Y_plt)
    plt.ylim(ymin=0, ymax=9)
    plt.xlim(xmin=0, xmax=6)
    plt.text(X_plt[-1], Y_plt[-1],
             ' y = ' +
             str('%.2f' %ridge_coeff) +
             ' * x' +
             ' + ' +
             str('%.2f' %ridge_intercept) +
             '      for \u03BB or \u03B1 = ' + str(l), fontsize=12)
```

```python
# Create subplots for Ridge Regression lines in one figure
for i, l in enumerate(ridge_l):
    ridge = Ridge(alpha=l)
    ridge.fit(X, Y)

    ridge_coeff = ridge.coef_
    ridge_intercept = ridge.intercept_
    Y_plt = ridge.predict(pd.DataFrame(X_plt))

    plt.figure(1)
    plt.subplot(2,2,i+1)
    plt.plot(X_plt, Y_plt)
    plt.ylim(ymin=0, ymax=9)
    plt.xlim(xmin=0, xmax=6)
    plt.title(' y = ' +
             str('%.2f' %ridge_coeff) +
             ' * x' +
             ' + ' +
             str('%.2f' %ridge_intercept) +
             '      for \u03BB or \u03B1 = ' + str(l), fontsize=12)
    plt.tight_layout()
```

```python
# Find pairs of features with high correlation (| correlation| > 0.8)
print("\nFeature pairs with high correlation (|correlation| > 0.8)\n")
high_corr = []
for i in range(len(correlation_matrix.columns)):
  for j in range(i + 1, len(correlation_matrix.columns)):
    corr_value = correlation_matrix.iloc[i, j]
    if abs(corr_value) > 0.8:
      high_corr.append((correlation_matrix.columns[i], correlation_matrix.columns[j], corr_value))

# Sort pairs by absolute value of correlation in descending order
high_corr = sorted(high_corr, key=lambda x: abs(x[2]), reverse=True)

# Output sorted pairs
if high_corr:
  for feat1, feat2, corr in high_corr:
    print(f"{feat1} \tAND \t{feat2}\n\tcorrelation = {corr:.4f}\n")
else:
  print("No high correlation (|correlation| > 0.8) was found.")
```

```python
# Divide the data into groups by Bankrupt? (0 и 1)
group_0 = df[df['Bankrupt?'] == 0]  # Not bankrupt
group_1 = df[df['Bankrupt?'] == 1]  # Bankrupt
```

```python
# Perform a t-test for each feature
print("Results of t-test:")
for feature in key_features:
  # Select data for each group
  values_0 = group_0[feature]
  values_1 = group_1[feature]

  # Checking if there is data in both groups
  if len(values_0) > 0 and len(values_1) > 0:
    t_stat, p_value = ttest_ind(values_0, values_1, equal_var=False)  # Using Welch's t-test.
    print(f"\nFeature: {feature}")
    print(f"t-statistics: {t_stat:.4f}")
    print(f"p-value: {p_value:.4f}")
    if p_value < 0.05:
      print("Result: Mean values differ (p < 0.05)")
    else:
      print("Result: No significant differences (p >= 0.05)")
  else:
    print(f"\nFeature: {feature}")
    print("Insufficient data for t-test (one of the groups is empty)")
```

```python
# ANOVA for each feature
print("ANOVA results for key features:")
for feature in key_features:
  # Select data for each group
  values_0 = group_0[feature]
  values_1 = group_1[feature]

  # Checking if there is data in both groups
  if len(values_0) > 0 and len(values_1) > 0:
    f_stat, p_value = f_oneway(values_0, values_1)
    print(f"\nFeature: {feature}")
    print(f"F-statistics: {f_stat:.4f}")
    print(f"p-value: {p_value:.4f}")
    if p_value < 0.05:
      print("Result: Mean values differ (p < 0.05)")
    else:
      print("Result: No significant differences (p >= 0.05)")
  else:
    print(f"\nFeature: {feature}")
    print("Insufficient data for ANOVA (one of the groups is empty)")
```

```python
X = df.drop('Bankrupt?', axis=1)  # Все столбцы, кроме Bankrupt?
y = df['Bankrupt?']  # Целевая переменная

print("Determination of the top 10 traits using ANOVA:")
anova_results = []
for feature in X.columns:
    # Separating the data into Bankrupt classes?
    group_0 = df[df['Bankrupt?'] == 0][feature]
    group_1 = df[df['Bankrupt?'] == 1][feature]

    # Checking if there is data in both groups
    if len(group_0) > 0 and len(group_1) > 0:
        f_stat, p_value = f_oneway(group_0, group_1)
        anova_results.append((feature, f_stat, p_value))
    else:
        anova_results.append((feature, 0, 1.0))  # If no data are available, assign low significance

# Sort by F-statistic (descending) and select top 10
anova_results = sorted(anova_results, key=lambda x: x[1], reverse=True)[:10]

# ANOVA results
print("\nTop 10 features by ANOVA (descending F- statistics):")
for feature, f_stat, p_value in anova_results:
    print(f"Feature: {feature}, F-statistics: {f_stat:.4f}, p-value: {p_value:.4f}")

# --- Feature Importance using Random Forest ---
print("\nIdentifying the top 10 features using Random Forest Feature Importance:")
# Creating and training a Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X, y)

# Getting the importance of attributes
importances = rf_model.feature_importances_
feature_importance = list(zip(X.columns, importances))

# Sort by importance (descending) and select the top 10
feature_importance = sorted(feature_importance, key=lambda x: x[1], reverse=True)[:10]

# Display Feature Importance results
print("\nTop 10 features by Random Forest Feature Importance:")
for feature, importance in feature_importance:
    print(f"Feature: {feature}, Importance: {importance:.4f}")
```

```python
le = LabelEncoder()
df['Loan_Status'] = le.fit_transform(df['Loan_Status'])  # Y=1, N=0
df['Gender'] = le.fit_transform(df['Gender'])  # Male=1, Female=0
```

```python
# Splitting the dataset into features (X) and target (y)
X = df.drop(['Loan_ID', 'Loan_Status'], axis=1)  # Dropping Loan_ID and target column
y = df['Loan_Status']  # Target variable

# One-hot encoding for categorical variables
X = pd.get_dummies(X, drop_first=True)

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Standardizing the numerical features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Importing ensemble classifiers

# Defining individual classifiers

log_clf = LogisticRegression(random_state=42)
dt_clf = DecisionTreeClassifier(random_state=42)
xgb_clf = XGBClassifier(random_state=42, eval_metric='logloss')
cat_clf = CatBoostClassifier(verbose=0, random_state=42)

# Voting Classifier
voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('dt', dt_clf), ('xgb', xgb_clf), ('cat', cat_clf)],
    voting='soft'
)

# Training the Voting Classifier
voting_clf.fit(X_train, y_train)

# Evaluating the model
y_pred = voting_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of Voting Classifier: {accuracy:.4f}")
```
