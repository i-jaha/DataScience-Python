Variable dictionary is listed below:
Manufacturer: manufacturer name
*   Model: model name
*   Displ: engine displacement, in litres
*   Year: year of manufacture
*   Cyl: number of cylinders
*   Trans: type of transmission
*   Drv: the type of drive train, where f = front-wheel drive, r = rear wheel drive, 4 = 4wd
*   Cty: city miles per gallon
*   Hwy: highway miles per gallon
*   Fl: fuel type
*   Class: "type" of car

Please, answer following questions:
1. Show relationship between highway and city miles per gallon.
  * Use scatter in matplotlib.
  * Interpret how are variables correlated according to scatter.
2. Show distributions and scatters between all variables.
  - While hue equals to type of drive train which variables have the lowest and the highest correlation?

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
cars = pd.read_excel('cars.xlsx')
cars.head()
````

````python
"""
Show relationship between highway and city miles per gallon.
Use scatter in matplotlib.
Interpret how are variables correlated according to scatter.
"""
plt.scatter(cars['cty'], cars['hwy'], alpha=0.6)
plt.title('Highway vs City Fuel Efficiency')
plt.xlabel('City MPG (Cty)')
plt.ylabel('Highway MPG (Hwy)')
plt.grid(True)
plt.show()
````

````python
cars.dtypes
````

````python
"""
Show distributions and scatters between all variables.
While hue equals to type of drive train which variables have the lowest and the highest correlation?
Find according to scatterplot.
"""
sns.pairplot(cars, hue='drv')
plt.show()
````

````python
"""
Which type of car is most frequent in dataset? Show by using countplot.
"""
most_frequent_class = cars['class'].value_counts().idxmax()
print(f"Most frequent car: {most_frequent_class}")

sns.countplot(cars, x='class')
plt.xticks(rotation=45)
plt.title("Quantity of cars by class")
plt.tight_layout()
plt.show()
````

````python
"""
Display number of cylinders for each drive train in bar chart.
Which drive train is the most frequent?
"""
most_frequent_train = cars['drv'].value_counts().idxmax()
print(f"Most frequent car: {most_frequent_train}")

sns.countplot(cars, x='drv', hue='cyl')
plt.show()
````

````python
"""
Visualize engine displacement by each class using boxplot.
Do the same thing in violinplot.
"""
fig, axes = plt.subplots(2, 1, figsize=(8, 12))

sns.boxplot(data=cars, x='class', y='displ', ax=axes[0])
axes[0].set_title('Engine Displacement by Class (Boxplot)')
axes[0].tick_params(axis='x', rotation=45)

sns.violinplot(data=cars, x='class', y='displ', ax=axes[1])
axes[1].set_title('Engine Displacement by Class (Violinplot)')
axes[1].tick_params(axis='x', rotation=45)

plt.show()
````
